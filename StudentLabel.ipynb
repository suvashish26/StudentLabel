```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# independent
# Features: [study_hours, sleep_hours, attendance%, previous_grades(out of 100), homework_hours]
X = np.array([
    [2, 6, 70, 50, 1],    # student 1
    [4, 7, 80, 75, 2],    # student 2
    [1, 5, 60, 40, 0.5],  # student 3
    [5, 8, 90, 85, 3],    # student 4
    [3, 6, 75, 60, 1.5],  # student 5
    [6, 7, 95, 90, 4],    # student 6
    [2, 5, 65, 55, 1],    # student 7
    [4, 6, 85, 80, 2.5],  # student 8
])

# dependent
# Labels: Pass or Fail
Y = np.array(['Fail', 'Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Fail', 'Pass'])

#training model using RandomForestClassifier
clf= RandomForestClassifier(n_estimators=10)
clf= clf.fit(X,Y)

#predection
prediction = clf.predict([[8,1,90,90,2]])
print(prediction)
```
